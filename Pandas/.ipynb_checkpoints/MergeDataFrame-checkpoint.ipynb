```python
import pandas as pd
import numpy as np
```

## Cadastro da loja A

```python
cad_a = {
            "Id":     ['AA2930', 'BB4563', 'CC2139', 'DE2521', 'GT3462', 'HH1158'],
            "Nome":   ['Victor', 'Danielle', 'Bruna', 'Carlos', 'Ricardo', 'Maria'],
            "Idade":  [20,35,40,54,30,27],
            "CEP":    ['00092-029','11111-111','22222-888','00000-999','88888-111','77777-666']
}
cad_a = pd.DataFrame(cad_a)
cad_a
```

## Cadastro da loja B

```python
cad_b = {
            "Id":     ['CC9999', 'EF4488', 'DD9999', 'GT3462', 'HH1158'],
            "Nome":   ['Marcos', 'Patricia', 'Ericka', 'Ricardo', 'Maria'],
            "Idade":  [19,30,22,30,27],
            "CEP":    ['00092-029','11111-111','11111-888', '88888-111','77777-666']
}
cad_b = pd.DataFrame(cad_b)
cad_b
```

## Registro de compras de todas unidades

```python
compras = {
            "Id" :    ['AA2930', 'EF4488', 'CC2139', 'EF4488', 'CC9999', 'AA2930', 'HH1158', 'HH1158'],
            "Data":   ['2019-01-01', '2019-01-30', '2019-01-30', '2019-02-01', '2019-02-20', '2019-03-15', '2019-04-01', '2019-04-10'],
            "Valor":  [200,100,40,150,300,25,50,500]
}
compras = pd.DataFrame(compras)
compras
```

```python
# pd.merge(tabela_da_esquerda, tabela_da_direita, on="Coluna_conincidente", how="left|right|inner|outer")
```

## Vamos verificar os clientes da loja 'A'  que também fizeram compras na loja 'B'

```python
pd.merge(cad_a, cad_b, on="Id", how="inner")
```

> As colunas com _x são da tabela da esquerda e com o _y são da tabela da direita

### Vamos unificar as colunas repetidas

```python
pd.merge(cad_a, cad_b[['Id', 'Idade', 'CEP']], on="Id", how="inner")
```

> Os sufixos x e y não são o que desejamos mostrar

### Mudando os Sufixos

```python
pd.merge(cad_a, cad_b[['Id', 'Idade', 'CEP']], on="Id", how="inner", suffixes=('_A', '_B'))
```

## Vamos realizar a junção total das bases - Full Join
#### Vamos abordar uma maneira de realizar esta junção usando o concat

```python
lojas = pd.concat([cad_a, cad_b], ignore_index=True)
lojas
```

#### Porém desta forma mostramos dados duplicados
> Ricardo e Maria
### Então vamos remover os duplicados pelo id:

```python
lojas.drop_duplicates(subset="Id")
```

## Realizando operação Left

```python
esquerda = pd.merge(cad_a,compras, on="Id", how="left")
esquerda
```

```python
esquerda.groupby(['Id', 'Nome'])['Valor'].sum()
```

## Realizando operação Outer

```python
pd.merge(cad_a, cad_b, on ="Id", how="outer", indicator=True) # O parametro indicator=True vai mostrar quais dados estão na esquerda, quais estão na direita e quais estão nas duas
```
